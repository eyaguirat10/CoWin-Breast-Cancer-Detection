# I. Business Understanding

## **Problem Statement**

Breast cancer diagnosis often depends on manual interpretation of medical measurements, which can lead to delays or inconsistencies in detecting malignant tumors. Hospitals need a reliable way to support clinicians with **fast, objective, and accurate assessments** based on patient data. By analyzing tumor characteristics extracted from the **Wisconsin Diagnostic Breast Cancer dataset**, this project seeks to build a system that improves early detection and reduces diagnostic errors.



---

## **Supervised Learning**

1.  ### Classification Objective

    * **Business Objective (BO):** Develop a reliable, accurate system for breast cancer detection.
   
    * **Data Science Objective (DSO):** Predict the diagnosis type — **M (Malignant)** or **B (Benign)**.

2.  ### Regression Objective
   
    * **Business Objective (BO):** Determine the size of a breast tumor using patient cell measurements to help doctors assess severity and plan treatment.

    * **Data Science Objective (DSO):** Predict the **exact size** of a breast tumor based on the patient’s cellular features to help assess severity and support treatment planning.

### **Business Success Metrics**

These metrics require true labels to evaluate prediction performance:

* **Accuracy:** Overall proportion of correct classifications. **Target $\ge$ 98%**
* **Precision:** Reliability of positive (malignant) predictions. **Target $\ge$ 97%**
* **F1-Score:** Balances recall and precision, especially with class imbalance. **Target $\ge$ 97%**
* **AUC-ROC:** Measures ability to discriminate between classes. **Target $\ge$ 0.99**

---

## **Unsupervised Learning**

* **Business Objective (BO):** Identify and group types of breast cancer cases for potential subtype discovery or anomaly detection.
* **Data Science Objective (DSO):** Cluster and classify diagnosis patterns without predefined labels.

### **Business Success Metrics**

Since labels are not available, success is measured differently:

* **Silhouette Score:** Measures how well-separated clusters are.
* **Davies-Bouldin Index / Calinski-Harabasz Index:** Assess cluster compactness and separation.
* **Domain Interpretability:** Clusters should be medically meaningful (e.g., malignant vs. benign enrichment).
